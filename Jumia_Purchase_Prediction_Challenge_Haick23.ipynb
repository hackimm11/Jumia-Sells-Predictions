```python
pivot_df['class'] = clust_labels
pivot_df
```

```python
pivot_df.sort_values(by = 'class', inplace = True)
pivot_df
```

```python
costumer_class = pivot_df['class']
costumer_class
```

```python
product_class = pivot_df.reset_index(drop = True)
product_class.head(20)
```

```python
class_df = product_class.groupby('class').mean()
class_df.head()
```

```python
# class_df['num_clients']=counter
```

```python
# class_df.head()
```

```python
solution_dic_id = []
solution_dic_cat = []
```

```python
# class_df['num_clients'].loc[0]
```

```python
# class_df.columns
```

```python
# class_df['num_clients'].loc[costumer_class[costumer]]
```

```python
for costumer in costumer_class.index:
    class_cost = costumer_class[costumer]
    for product in class_df.columns:
        solution_dic_id.append("_".join([str(costumer),product]))
        if class_df.loc[class_cost][product] > 0.07:
            solution_dic_cat.append('buy')
        else:
            solution_dic_cat.append('not buy')
            
#         if class_df.loc[class_cost][product] > class_df['num_clients'].loc[class_cost]/10.0:
#             solution_dic_cat.append('buy')
#         else:
#             solution_dic_cat.append('not buy')
```

```python
len(solution_dic_cat)
```

```python
solution_df = pd.DataFrame({'id':solution_dic_id,'cat':solution_dic_cat})
solution_df.head()
```
